# Cosine similarity

```python
#Cosine Similarity funciton
def cosine_similarity(v1,v2):
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]
        y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/float(np.sqrt(sumxx*sumyy))
```

```python
d=list(dat['Retweet'])
for i in range(len(d)):
    if(d[i]!='-1'):
        s=str(d[i])
        #if s in u1:
        #    index = u1.index(s)
        #    print(index)
       
        rid=random.choice(userid)
        ind=userid.index(rid)
        rv=random.randrange(1,4,1)
        ind1=userid.index(u1[i])
        arr[ind1][ind]+=rv
```

```python
n=26
imatrix=np.zeros([n,n],dtype=float) #(u*u matrix) with interaction count
for i in range(len(arr)):
    mip=0
    bo=0
    for j in range(len(arr)):
        eidu=len(arr)-(list(arr[i]).count(0))
        eidv=len(arr)-(list(arr[j]).count(0))
        a1=copy.copy(arr[i])
        for k in range(len(arr)):
            if a1[k]>0:
                a1[k]=1
        a2=copy.copy(arr[j])
        for k in range(len(arr)):
            if a2[k]>0:
                a2[k]=1
        if eidu==0:
            mip=0
        else:
            mip=(len(arr)-list(a1 & a2).count(0))/float(eidu)
        if eidv==0:
            bo=0
        else:
            bo=(len(arr)-list(a1 & a2).count(0))/float(eidv)
        imatrix[i,j]=(0.5*mip)+(0.5*bo)
```

# Comprehensive influence

```python
#Comprehensive Influence
comp_inf=np.matrix(imatrix)+((0.5*np.matrix(ucos))+(0.5*np.matrix(kinf)))

comp_inf_sum=[]
for i in range(len(comp_inf)):
    comp_inf_sum.append(np.matrix.sum(comp_inf[i]))
```
